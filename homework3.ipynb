```python
# 2. Дан список повторяющихся элементов. Вернуть список с дублирующимися элементами. В результирующем списке не должно быть дубликатов.


from random import randint

LOWER_LIMIT = 0
UPPER_LIMIT = 10
COUNT = 20

orig_list = [randint(LOWER_LIMIT, UPPER_LIMIT) for _ in range(COUNT)]

print('Исходный список: ', orig_list)

uniq_elem = set(orig_list)

# print('Уникальные элементы', uniq_elem)

repeat_elem = set()
for elem in uniq_elem:
    if orig_list.count(elem) > 1:
        repeat_elem.add(elem)

print('Повторяющиеся элементы:', repeat_elem)
```

```python
# 3. В большой текстовой строке подсчитать количество встречаемых слов и вернуть 10 самых частых.
# Не учитывать знаки препинания и регистр символов. За основу возьмите любую статью из википедии или из документации к языку.


import string
from collections import Counter

source_text = '''
    Set types

    These represent unordered, finite sets of unique, immutable objects. As such, they cannot be indexed by any subscript.
    However, they can be iterated over, and the built-in function len() returns the number of items in a set. Common uses
    for sets are fast membership testing, removing duplicates from a sequence, and computing mathematical operations such
    as intersection, union, difference, and symmetric difference.

    For set elements, the same immutability rules apply as for dictionary keys. Note that numeric types obey the normal rules
    for numeric comparison: if two numbers compare equal (e.g., 1 and 1.0), only one of them can be contained in a set.

    There are currently two intrinsic set types:

    Sets
    These represent a mutable set. They are created by the built-in set() constructor and can be modified afterwards
    by several methods, such as add().

    Frozen sets
    These represent an immutable set. They are created by the built-in frozenset() constructor. As a frozenset is immutable
    and hashable, it can be used again as an element of another set, or as a dictionary key.
    '''

table = str.maketrans("", "", string.punctuation)
clear_text = source_text.translate(table).lower()

words = list(clear_text.split())

word_counts = Counter(words)
most_common_words = word_counts.most_common(10)

print('10 самых часто используемых слов: ',*most_common_words)
```

```python
# 4. Создайте словарь со списком вещей для похода в качестве ключа и их массой в качестве значения.
# Определите какие вещи влезут в рюкзак передав его максимальную грузоподъёмность. Достаточно вернуть один допустимый вариант.
# *Верните все возможные варианты комплектации рюкзака.


def pack_combination(items, max_weight):
    combination = []

    possible_combinations = []

    def backtrack(current_items, current_weight, index):
        if current_weight <= max_weight and current_items not in possible_combinations:
            possible_combinations.append(current_items)

        for i in range(index + 1, len(items)):
            new_item = items[i]
            new_weight = current_weight + new_item[1]

            if new_weight <= max_weight:
                backtrack(current_items + [new_item], new_weight, i)

    backtrack([], 0, -1)

    return possible_combinations


items = [
    ('Палатка', 5),
    ('Котелок', 3),
    ('Аптечка', 2),
    ('Продукты', 10),
    ('Фонарик', 1),
    ('Топор', 8),
    ('Спальный мешок', 7)
    ]
max_weight = 10

all_combinations = pack_combination(items, max_weight)
for combination in all_combinations:
    print('Возможная комплектация:', *combination if combination != [] else 'Пусто', ', Общий вес:', sum(item[1] for item in combination))

```
