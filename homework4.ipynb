```python
# 1. Напишите функцию для транспонирования матрицы


def mattrix_transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]


def mattrix_print(matrix):
    max_len = max(len(str(num)) for row in matrix for num in row)
    for row in matrix:
        for elem in row:
            print(f'{elem:{max_len}}', end=' ')
        print()


mattrix = [
    ['one', 'two', 'three', 'four'],
    [10, 20, 30, 40],
    [100, 200, 300, 400]
    ]

print('Исходная матрица:')
mattrix_print(mattrix)

tran_mattrix = mattrix_transpose(mattrix)

print('Транспонированная матрица:')
mattrix_print(tran_mattrix)
```

```python
# 2. Напишите функцию принимающую на вход только ключевые параметры и возвращающую словарь, где ключ — значение переданного аргумента,
# а значение — имя аргумента. Если ключ не хешируем, используйте его строковое представление.


def args_to_dict(**kwargs):
    result = {}
    for key, value in kwargs.items():
        try:
            result[value] = key
        except TypeError:
            result[str(value)] = key
    return result


print(args_to_dict(a=1, b=2))
print(args_to_dict(a=1, b=2, c=(1,2,3)))
```

```python
# 3. Возьмите задачу о банкомате из семинара 2. Разбейте её на отдельные операции — функции.
# Дополнительно сохраняйте все операции поступления и снятия средств в список.

# Напишите программу банкомат.
# Начальная сумма равна нулю
# Допустимые действия: пополнитть, снять, выйти
# Сумма пополнения и снятия кратны 50 у.е.
# Процент за снятие - 1,5% от суммы снятия, но не менее 30 и не более 600 у.е.
# После каждой третьей операции пополнения или снятия начисляютсч проценты - 3%
# Нельзя снять больше, чем на счёте
# При превышении суммы в 5 млн, вычитается налог на богатство 10% перед каждой операцией, даже ошибочной
# любое действие выводиь сумму денег

import decimal

CMD_DEPOSIT = 'п'
CMD_WITHDRAW = 'с'
CMD_EXIT = 'в'
RICHNESS_SUM = decimal.Decimal(5_000_000)
RICHNESS_TAX = decimal.Decimal(10) / decimal.Decimal(100)
WITHDRAW_PERCENT = decimal.Decimal(1.5) / decimal.Decimal(100)
ADD_PERCENT = decimal.Decimal(3) / decimal.Decimal(100)
MULTIPLICITY = 50
MIN_REMOVAL = 30
MAX_REMOVAL = 600
COUNT_OPER = 3


def amount_check_():
    _amount = 1
    while _amount % MULTIPLICITY != 0:
        _amount = int(input(f'Введите сумму в {MULTIPLICITY} у.е.: '))
    return _amount


def cmd_depostit_(_account, _amount, _count):
    _account += _amount
    _count += 1
    print(f'Пополнение карты на {_amount} у.е.\n',
          f'Итого на карте {_account} у.е.')
    return _account, _amount, _count


def cmd_withdraw_(_account,_amount,_count):
    if _account >= _amount + withdraw_tax_(_amount):
        _account -= (_amount + withdraw_tax_(_amount))
        _count += 1
        transactions.append(('Комиссия',withdraw_tax_(_amount),int(account)))
        print(f'Снятие с карты {_amount} у.е.\n',
              f'Комиссия за снятие {withdraw_tax_(_amount)} у.е.\n',
              f'На карте осталось {_account} у.е.')
    else:
        print(f'Недостаточно денег для выполнения операции\n',
              f'На карте {_account:.2} у.е.')
    return _account,_amount,_count


def withdraw_tax_(_amount):
    withdraw_tax = _amount * WITHDRAW_PERCENT
    withdraw_tax = (MIN_REMOVAL if withdraw_tax < MIN_REMOVAL else
                    MAX_REMOVAL if withdraw_tax > MAX_REMOVAL else withdraw_tax)
    return withdraw_tax


def richness_tax_(_account):
    percent = _account * RICHNESS_TAX
    _account -= percent
    print(f'Удержан налог на богатство {RICHNESS_TAX * 100}% в размере {percent} у.е.\n',
          f'Итого на карте осталось {_account:.2f} у.е.')
    return percent, _account


def bonus_percent_(_account):
    bonus_percent = _account * ADD_PERCENT
    _account += bonus_percent
    print(f'На счет начислен бонус {ADD_PERCENT * 100}% в размере {bonus_percent} у.е.\n',
          f'Итого на карте {_account:.2f} у.е.')
    return bonus_percent, _account


def cmd_exit_():
    print(f'Возьмите карту, на которой {account:.2f} у.е.')
    return


account = decimal.Decimal(0)
count = 0
transactions = []

while True:
    if account > RICHNESS_SUM:
        tax, account = richness_tax_(account)
        transactions.append(('Налог',round(tax,2),round(account,2)))
    command = input(f'Пополнить - {CMD_DEPOSIT}, Снять - {CMD_WITHDRAW}, Выйти - {CMD_EXIT}: ').lower()
    if command == CMD_EXIT:
        cmd_exit_()
        break
    if command in (CMD_DEPOSIT, CMD_WITHDRAW):
        amount = amount_check_()
        if command == CMD_DEPOSIT:
            account, amount, count = cmd_depostit_(account,amount,count)
            transactions.append(('Приход',round(amount,2),round(account,2)))
        elif command == CMD_WITHDRAW:
            account, amount, count = cmd_withdraw_(account,amount,count)
            transactions.append(('Расход',round(amount,2),round(account,2)))
        if count > 0 and count % COUNT_OPER == 0:
            bonus, account = bonus_percent_(account)
            transactions.append(('Бонус',round(bonus,2),round(account,2)))
    else:
        print('Неверная команда')


```
